# Bar char for each feature

```python
# Distribution of age
sns.countplot(data=df, x='Age')
plt.title(f'Distribution of Age')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Gender
sns.countplot(data=df, x='Gender', palette={'0': 'blue', '1': 'red'})
plt.title(f'Distribution of Gender')
plt.ylabel('Count')
handles = [
    plt.Line2D([0], [0], color='blue', lw=4),
    plt.Line2D([0], [0], color='red', lw=4)
]
labels = ['Male (0)', 'Female (1)']
plt.legend(handles=handles, labels=labels, title='Gender', loc='best')

plt.show()
```

```python
# Distribution of Ethnicity
sns.countplot(data=df, x='Ethnicity')
plt.title(f'Distribution of Ethnicity')
plt.ylabel('Count')
handles = [
    plt.Line2D([0], [0], lw=4),
    plt.Line2D([0], [0], lw=4),
    plt.Line2D([0], [0], lw=4),
    plt.Line2D([0], [0], lw=4),
]
labels = ['Caucasian (0)', 'African American (1)', 'Asian (2)', 'Other (3)']
plt.legend(handles=handles, labels=labels, title='Ethnicity', loc='best')

plt.show()
```

```python
# Distribution of ParentalEducation
sns.countplot(data=df, x='ParentalEducation')
plt.title(f'Distribution of Parental Education')
plt.ylabel('Count')
handles = [
    plt.Line2D([0], [0], lw=4),
    plt.Line2D([0], [0], lw=4),
    plt.Line2D([0], [0], lw=4),
    plt.Line2D([0], [0], lw=4),
]
labels = ['None (0)', 'High School (1)', 'Some College (2)', 'Bachelor (3)', 'Higher (4)']
plt.legend(handles=handles, labels=labels, title='Ethnicity', loc='best')

plt.show()
```

```python
# Distribution of GPA
sns.histplot(df['GPA'], bins=4, kde=True, color='blue')
plt.title('Distribution of GPA')
plt.xlabel('GPA)')
plt.ylabel('Count')
plt.show()

# check how many student has GPA lower 2.0
print(df[df['GPA'] <= 2.0].count) # => 1274 / 2392
```

```python
# GradeClass: Classification of students' grades based on GPA:
# 0: 'A' (GPA >= 3.5)
# 1: 'B' (3.0 <= GPA < 3.5)
# 2: 'C' (2.5 <= GPA < 3.0)
# 3: 'D' (2.0 <= GPA < 2.5)
# 4: 'F' (GPA < 2.0)
# Distribution of GradeClass 
plt.figure(figsize=(5, 5))
sns.countplot(data=df, x='GradeClass')
plt.title(f'Distribution of GradeClass')
plt.ylabel('Count')
plt.show()
# Distribution of GradeClass with historygram
sns.histplot(df['GradeClass'], bins=5, kde=True, color='blue')
plt.title('Distribution of GradeClass')
plt.xlabel('GradeClass)')
plt.ylabel('Count')
plt.show()

```
